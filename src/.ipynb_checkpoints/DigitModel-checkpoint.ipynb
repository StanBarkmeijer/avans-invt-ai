```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train= x_train / 255.0
x_test = x_train / 255.0

x_train, x_valid, y_train, y_valid = train_test_split(
    x_train,
    y_train, 
    test_size=0.2, 
    random_state=42
)

x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)
x_valid = x_valid.reshape(-1, 28*28)
```

```python
y_train_binary = (y_train == 1).astype(int)
y_test_binary = (y_test == 1).astype(int)
y_test_binary = np.repeat(y_test_binary, 6, axis=0)
y_valid_binary = (y_valid == 1).astype(int)
```

## 1 Layer Model

```python
model = tf.keras.models.Sequential([
 tf.keras.layers.Dense(128, activation='relu'),
 tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss = tf.keras.losses.BinaryCrossentropy(from_logits=False), 
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
history = model.fit(
    x_train, 
    y_train_binary, 
    epochs=10,
    validation_data = (x_valid, y_valid_binary),
    callbacks=[early_stopping]
)
```

```python
results = []

result = {
 'layers': 1,
 'activation': 'relu, sigmoid',
 'history': history,
 'loss': model.evaluate(x_test, y_test_binary, verbose=2)[0],
 'accuracy': model.evaluate(x_test, y_test_binary, verbose=2)[1],
 'auc': model.evaluate(x_test, y_test_binary, verbose=2)[2]
}

results.append(result)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
print(result)
```

## 3 Layer model

```python
model = tf.keras.models.Sequential([
 tf.keras.layers.Dense(128, activation='relu'),
 tf.keras.layers.Dense(64, activation='relu'),
 tf.keras.layers.Dense(32, activation='relu'),
 tf.keras.layers.Dense(1, activation='sigmoid') 
])

model.compile(
    optimizer='adam',
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=False),
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

history = model.fit(
    x_train, 
    y_train_binary, 
    epochs=10,
    validation_data=(x_valid, y_valid_binary), 
    callbacks=[early_stopping]
)
```

```python
result = {
 'layers': 3,
 'activation': 'relu',
 'history': history,
 'loss': model.evaluate(x_test, y_test_binary, verbose=2)[0],
 'accuracy': model.evaluate(x_test, y_test_binary, verbose=2)[1],
 'auc': model.evaluate(x_test, y_test_binary, verbose=2)[2]
}

results.append(result)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
print(result)
```

## Categorical

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=3)

y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
y_valid = to_categorical(y_valid, num_classes=10)
```

```python
model = tf.keras.models.Sequential([
 tf.keras.layers.Dense(128, activation='relu'),
 tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

history = model.fit(x_train, y_train, epochs=10,
validation_data=(x_valid, y_valid), callbacks=[early_stopping])
```

```python
result = {
 'layers': 1,
 'activation': 'relu',
 'history': history,
 'loss': model.evaluate(x_test, y_test, verbose=2)[0],
 'accuracy': model.evaluate(x_test, y_test, verbose=2)[1],
 'auc': model.evaluate(x_test, y_test, verbose=2)[2]
}

results.append(result)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
print(result)
```
