```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
import nltk

from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
```

```python
afinn_path = '../data/AFINN.json'
tweets_path = '../data/twitter_training.csv'
tweets_val_path = '../data/twitter_validation.csv'

afinn = pd.read_json(afinn_path, typ='series')
train = pd.read_csv(tweets_path, sep=',', names=['ID', 'Entity', 'Sentiment', 'Content'])
val = pd.read_csv(tweets_val_path, sep=',', names=['ID', 'Entity', 'Sentiment', 'Content'])
```

```python
train.dropna(inplace=True)
train.drop_duplicates(inplace=True)
train.drop(columns=['Sentiment'], inplace=True)
```

```python
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'http\S+', '', text)
        text = re.sub(r'@\w+','', text)
        text = re.sub(r'#\w+','', text)
        text = re.sub(r'[^a-zA-Z\s]','', text)
        text = text.lower()
    return text

def remove_emojis(text):
    if isinstance(text, str):
        text = re.sub(r'\\x\S+', '', text)
    return text

train['Content'] = train['Content'].apply(clean_text)
train['Content'] = train['Content'].apply(remove_emojis)

val['Content'] = val['Content'].apply(clean_text)
val['Content'] = val['Content'].apply(remove_emojis)
```

```python
def calculate_sentiment(text):
    if isinstance(text, str):
        words = text.split()
        sentiment = 0
        for word in words:
            if word in afinn:
                sentiment += afinn[word]
        return sentiment
    return 0

train['Sentiment'] = train['Content'].apply(calculate_sentiment)
train['SentimentRating'] = train['Sentiment'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))
train['WordCount'] = train['Content'].apply(lambda x: len(x.split()))

val['Sentiment'] = val['Content'].apply(calculate_sentiment)
val['SentimentRating'] = val['Sentiment'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))
val['WordCount'] = val['Content'].apply(lambda x: len(x.split()))
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

train = remove_outlier(train, 'WordCount')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words and word.isalpha()]
    return " ".join(filtered_tokens)

train['preprocessed_text'] = train['Content'].apply(lambda x: preprocess(x))
val['preprocessed_text'] = val['Content'].apply(lambda x: preprocess(x))
```

```python
X_train ,X_test , y_train, y_test = train_test_split(
    train[['preprocessed_text']],
    train[['SentimentRating']],
    test_size=0.2,
    random_state=42
)
```

```python
vectorizer= TfidfVectorizer()

X_train_vect= vectorizer.fit_transform(X_train['preprocessed_text'])
X_test_vect= vectorizer.transform(X_test['preprocessed_text'])

y_train=y_train['SentimentRating'].map({"positive":0,"negative":1,"neutral":2})
y_test=y_test['SentimentRating'].map({"positive":0,"negative":1,"neutral":2})
```

```python
validation_X = vectorizer.transform(val['preprocessed_text'])
validation_y = val['SentimentRating'].map({"positive":0,"negative":1,"neutral":2})
```

```python
model = MultinomialNB()
model.fit(X_train_vect,y_train)
y_predict= model.predict(X_test_vect)
print(classification_report(y_test,y_predict))
```

```python
model = RandomForestClassifier()
model.fit(X_train_vect,y_train)
y_predict= model.predict(X_test_vect)

print(classification_report(y_test,y_predict))
```

```python
model_extra = ExtraTreesClassifier()
model_extra.fit(X_train_vect,y_train)
y_predict= model_extra.predict(X_test_vect)
print(classification_report(y_test,y_predict))
```
